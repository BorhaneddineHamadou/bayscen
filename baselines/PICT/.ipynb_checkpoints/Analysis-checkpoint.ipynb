## Here we will convert the files to Excel format so that they are consistent with the other files

```python
import pandas as pd

# Input CSV file
input_csv = "scenarios_scenario1_2w.csv"

# Output Excel file
output_excel = "PICT_2w_scenario1.xlsx"

# Read CSV using tab as separator
df = pd.read_csv(input_csv, sep='\t')

# Save to Excel
df.to_excel(output_excel, index=False)

print(f"File saved as {output_excel}")
```
