```python
import json

# Load the run results
with open("run_results.json", "r") as f:
    run_results = json.load(f)

# Load the min_ttc data
with open("min_ttc_log.json", "r") as f:
    ttc_data = json.load(f)

# Check that both files have the same number of entries
assert len(run_results) == len(ttc_data), "Mismatch in entry count between files"

# Create the new TTC list
ttc_new = []
for run, ttc in zip(run_results, ttc_data):
    new_entry = {"min_ttc": 0.0 if run["collision_occurred"] else ttc["min_ttc"]}
    ttc_new.append(new_entry)

# Save the new TTC values to a file
with open("TTC_new.json", "w") as f:
    json.dump(ttc_new, f, indent=2)

print("TTC_new.json created successfully.")
```
